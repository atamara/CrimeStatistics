```python
import pandas as pd
```

```python
#Calling in Hate Crime dataset
HateCrime2019 = pd.read_csv("Data Files/Hate_Crime_Incidents_per_Bias_Motivation_and_Quarter_by_State_Federal_and_Agency_2019.csv")
HateCrime2019.head()
```

```python
#Calling in Census dataset
USCensus2020 = pd.read_csv("Data Files/2020 US Census - State Racial Demographics.csv")
USCensus2020.head()
```

```python
#Data Wrangling
#Headers need to be cleaned up (remove spaces and symbols, and simplify/clarify)
HateCrime2019.rename(columns={'Agency Type' : 'AgencyType', 'Race/\nEthnicity/\nAncestry' : 'Race', 
                              'Sexual\norientation' : 'SexualOrientation', 'Gender\nidentity' : 'GenderIdentity', 
                              '1st\nquarter' : '1stQtr', '2nd\nquarter' : '2ndQtr', '3rd\nquarter' : '3rdQtr', 
                              '4th\nquarter' : '4thQtr'}, inplace=True)
USCensus2020.rename(columns={'% White' : 'WhitePct', '% Black' : 'BlackPct', '% American Indian and Alaska Native' : 
                             'AmerIndianOrAlaskaNativePct', '% Asian' : 'AsianPct', 
                             '% Native Hawaiian and Other Pacific Islander' : 'NativeHawaiiOrOtherPacificIslePct', 
                             '% Other Race' : 'OtherRacePct', '% Two or More Races' : 'TwoOrMoreRacePct', '% Hispanic' : 
                             'HispanicPct'}, inplace=True)
```

```python
#Viewing headers to confirm code worked
HateCrime2019.head()
```

```python
USCensus2020.head()
```

```python
#Since we are focusing our questions on the state level only, breakdown according to local agency is not needed.
#Removing the "AgencyType" and "Agency" columns to simplify the "HateCrime2019" dataset, and viewing to confirm code worked.
del HateCrime2019['AgencyType']
HateCrime2019.head()
```

```python
del HateCrime2019['Agency']
HateCrime2019.head()
```

```python
#Want to see all the data to get a sense of any issues with the dataset.
pd.options.display.max_rows = None
print(HateCrime2019)
```

```python
#Since we are focusing our questions on the state level only,
#we will remove any rows from the "HateCrime2019" dataset where the State value is "Federal"
HateCrime2019 = HateCrime2019[HateCrime2019.State != 'Federal']
```

```python
#There are a number of NaN values in "HateCrime2019" dataset. We can't use rows where state data is missing, so will drop those.
HateCrime2019 = HateCrime2019[HateCrime2019['State'].notna()]
```

```python
#Viewing to confirm code worked.
pd.options.display.max_rows = None
print(HateCrime2019)
```

```python
#Review of the file makes clear that the other NaN values all fall within the 1stQtr, 2ndQtr, 3rdQtr, 4thQtr, and
#Population columns. A review of the data across all columns makes clear that all NaN in Qtr columns should be 0 (total across
#Race, Religion, SexualOrientation, Disability, Gender, GenderIdentity should = total across Qtrs).
#As for NaN in Population column, no state is missing the majority of its population data. Considering our questions, feel it is
#more important to keep crime data and imperfect population data than to drop crime data because population data is missing.
#Have chosen to replace remaining NaN values with 0.
HateCrime2019_NoNA = HateCrime2019.fillna(0)
```

```python
print(HateCrime2019_NoNA)
```

```python
#Now need to collapse all rows of "HateCrime2019_NoNA" dataset for the same state into one row for ease of working with dataset.
#Since questions focused on the state level, do not need data broken out by local agency.
#This will also make it possible to combine with "USCensus2020" dataset which is one row per state.
```

```python
HateCrime2019_NoNA1 = HateCrime2019_NoNA.groupby(['State']).agg({'Race': ['sum']}).reset_index()
```

```python
#Verifying aggregate worked
print(HateCrime2019_NoNA1)
```

```python
#Aggregated all other columns into their own tables
HateCrime2019_NoNA3 = HateCrime2019_NoNA.groupby(['State']).agg({'SexualOrientation': ['sum']}).reset_index()
HateCrime2019_NoNA4 = HateCrime2019_NoNA.groupby(['State']).agg({'Disability': ['sum']}).reset_index()
HateCrime2019_NoNA5 = HateCrime2019_NoNA.groupby(['State']).agg({'Gender': ['sum']}).reset_index()
HateCrime2019_NoNA6 = HateCrime2019_NoNA.groupby(['State']).agg({'GenderIdentity': ['sum']}).reset_index()
HateCrime2019_NoNA7 = HateCrime2019_NoNA.groupby(['State']).agg({'1stQtr': ['sum']}).reset_index()
HateCrime2019_NoNA8 = HateCrime2019_NoNA.groupby(['State']).agg({'2ndQtr': ['sum']}).reset_index()
HateCrime2019_NoNA9 = HateCrime2019_NoNA.groupby(['State']).agg({'3rdQtr': ['sum']}).reset_index()
HateCrime2019_NoNA10 = HateCrime2019_NoNA.groupby(['State']).agg({'4thQtr': ['sum']}).reset_index()
HateCrime2019_NoNA11 = HateCrime2019_NoNA.groupby(['State']).agg({'Population': ['sum']}).reset_index()
```

```python
#Code did not work with Religion column - apparently these digits need to be converted from int to float
modelFeatures = ['Religion']
HateCrime2019_NoNA0 = HateCrime2019_NoNA[modelFeatures].astype(float) 
print(HateCrime2019_NoNA0)
```

```python
#To aggregate, I need the State information with the Religion information
HateCrime2019_NoNA00 = HateCrime2019_NoNA['State']
print(HateCrime2019_NoNA00)
```

```python
HateCrime2019_NoNA000 = pd.concat([HateCrime2019_NoNA00, HateCrime2019_NoNA0],axis=1)
print(HateCrime2019_NoNA000)
```

```python
#Now aggregate
HateCrime2019_NoNA2 = HateCrime2019_NoNA000.groupby(['State']).agg({'Religion': ['sum']}).reset_index()
```

```python
#Verifying conversion and aggregate worked
print(HateCrime2019_NoNA2)
```

```python
#Now that everything is set up for one row per state, I should be able to join all the separate tables into one dataset
#using the State column. Will test with the first 2
HateCrime2019_NoNA1and2 = pd.merge(HateCrime2019_NoNA1, HateCrime2019_NoNA2)
print(HateCrime2019_NoNA1and2)
```

```python
#Works! Now to join all the other tables created from original HateCrime2019 dataset
HateCrime2019_NoNA123 = pd.merge(HateCrime2019_NoNA1and2, HateCrime2019_NoNA3)
HateCrime2019_NoNA1to4 = pd.merge(HateCrime2019_NoNA123, HateCrime2019_NoNA4)
HateCrime2019_NoNA1to5 = pd.merge(HateCrime2019_NoNA1to4, HateCrime2019_NoNA5)
HateCrime2019_NoNA1to6 = pd.merge(HateCrime2019_NoNA1to5, HateCrime2019_NoNA6)
HateCrime2019_NoNA1to7 = pd.merge(HateCrime2019_NoNA1to6, HateCrime2019_NoNA7)
HateCrime2019_NoNA1to8 = pd.merge(HateCrime2019_NoNA1to7, HateCrime2019_NoNA8)
HateCrime2019_NoNA1to9 = pd.merge(HateCrime2019_NoNA1to8, HateCrime2019_NoNA9)
HateCrime2019_NoNA1to10 = pd.merge(HateCrime2019_NoNA1to9, HateCrime2019_NoNA10)
HateCrime2019_NoNATotal = pd.merge(HateCrime2019_NoNA1to10, HateCrime2019_NoNA11)
print(HateCrime2019_NoNATotal)
```

```python
#Now can combine with USCensus2020 dataset, which needs the United States row dropped to match up with HateCrimes dataset
print(USCensus2020)
```

```python
#Drop row 50 with the US data and leave only the state data
USCensus2020_StatesOnly = USCensus2020.drop([50], axis=0)
print(USCensus2020_StatesOnly)
```

```python
#Merge the HateCrimes and USCensus datasets
HateCrime2019_USCensus2020 = pd.merge(HateCrime2019_NoNATotal, USCensus2020_StatesOnly, on='State')
print(HateCrime2019_USCensus2020)
```

```python
#Drop the excess State column
HC2019_USC2020 = HateCrime2019_USCensus2020.drop(['State'], axis=1)
print(HC2019_USC2020)
```

```python
#Now that Data Wrangling is completed, will export dataframe to GitHub
HC2019_USC2020.to_csv(r'Data Files/HC2019_USC2020.csv')
```
