```python
import pandas as pd
```

```python
#Calling in Hate Crime dataset
HateCrime2019 = pd.read_csv("Data Files/Hate_Crime_Incidents_per_Bias_Motivation_and_Quarter_by_State_Federal_and_Agency_2019.csv")
HateCrime2019.head()
```

```python
#Calling in Census dataset
USCensus2020 = pd.read_csv("Data Files/2020 US Census - State Racial Demographics.csv")
USCensus2020.head()
```

```python
#Data Wrangling
#Headers need to be cleaned up (remove spaces and symbols, and simplify/clarify)
HateCrime2019.rename(columns={'Agency Type' : 'AgencyType', 'Race/\nEthnicity/\nAncestry' : 'Race', 
                              'Sexual\norientation' : 'SexualOrientation', 'Gender\nidentity' : 'GenderIdentity', 
                              '1st\nquarter' : '1stQtr', '2nd\nquarter' : '2ndQtr', '3rd\nquarter' : '3rdQtr', 
                              '4th\nquarter' : '4thQtr'}, inplace=True)
USCensus2020.rename(columns={'% White' : 'WhitePct', '% Black' : 'BlackPct', '% American Indian and Alaska Native' : 
                             'AmerIndianOrAlaskaNativePct', '% Asian' : 'AsianPct', 
                             '% Native Hawaiian and Other Pacific Islander' : 'NativeHawaiiOrOtherPacificIslePct', 
                             '% Other Race' : 'OtherRacePct', '% Two or More Races' : 'TwoOrMoreRacePct', '% Hispanic' : 
                             'HispanicPct'}, inplace=True)
```

```python
#Viewing headers to confirm code worked
HateCrime2019.head()
```

```python
USCensus2020.head()
```

```python
#Since we are focusing our questions on the state level only, breakdown according to local agency is not needed.
#Removing the "AgencyType" and "Agency" columns to simplify the "HateCrime2019" dataset, and viewing to confirm code worked.
del HateCrime2019['AgencyType']
HateCrime2019.head()
```

```python
del HateCrime2019['Agency']
HateCrime2019.head()
```

```python
#Want to see all the data to get a sense of any issues with the dataset.
pd.options.display.max_rows = None
print(HateCrime2019)
```

```python
#Since we are focusing our questions on the state level only,
#we will remove any rows from the "HateCrime2019" dataset where the State value is "Federal"
HateCrime2019 = HateCrime2019[HateCrime2019.State != 'Federal']
```

```python
#There are a number of NaN values in "HateCrime2019" dataset. We can't use rows where state data is missing, so will drop those.
HateCrime2019 = HateCrime2019[HateCrime2019['State'].notna()]
```

```python
#Viewing to confirm code worked.
pd.options.display.max_rows = None
print(HateCrime2019)
```

```python
#Review of the file makes clear that the other NaN values all fall within the 1stQtr, 2ndQtr, 3rdQtr, 4thQtr, and
#Population columns. A review of the data across all columns makes clear that all NaN in Qtr columns should be 0 (total across
#Race, Religion, SexualOrientation, Disability, Gender, GenderIdentity should = total across Qtrs).
#As for NaN in Population column, no state is missing the majority of its population data. Considering our questions, feel it is
#more important to keep crime data and imperfect population data than to drop crime data because population data is missing.
#Have chosen to replace remaining NaN values with 0.
HateCrime2019.fillna(0)
```

```python
#Now need to collapse all rows of "HateCrime2019" dataset for the same state into one row for ease of working with dataset.
#Since questions focused on the state level, do not need data broken out by local agency.
#This will also make it possible to combine with "USCensus2020" dataset which is one row per state.
```

```python
HateCrime2019.groupby(['State']).agg({'Race': ['sum']}).reset_index()
```
